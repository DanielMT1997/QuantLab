```python
import pandas as pd
import yfinance as yf
from pathlib import Path

print("Librerías cargadas correctamente")
```

```python
ticker = "SPY"
start_date = "1993-01-01"

project_root = Path.cwd().parent
raw_data_path = project_root / "data" / "raw"
raw_data_path.mkdir(parents=True, exist_ok=True)

output_file = raw_data_path / "SPY_daily.csv"

print("Carpeta de datos:", raw_data_path)
print("Archivo de salida:", output_file)
```

```python
# Descargamos datos originales
spy_raw = yf.download(
    ticker,
    start=start_date,
    auto_adjust=False,
    progress=False
)

spy_raw.head()
```

```python
raw_output_file = raw_data_path / "SPY_daily_raw.csv"
spy_raw.to_csv(raw_output_file)
```

```python
spy = spy_raw.copy()

if isinstance(spy.columns, pd.MultiIndex):
    spy.columns = spy.columns.get_level_values(0)

spy.head()
```

```python
processed_data_path = project_root / "data" / "processed"
processed_data_path.mkdir(parents=True, exist_ok=True)

processed_output_file = processed_data_path / "SPY_daily.csv"
spy.to_csv(processed_output_file)
```
